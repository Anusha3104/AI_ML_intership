```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1,2,3,4,5]
y=[22,30,12,45,25] # x and y should have same number of values
#plt.plot(x,y)

plt.plot(x,y,marker='s',linestyle=':',color='green', linewidth=2) # point markers: 'o','s','x',^,'*
                                        # linestyle: '-','--',":","-."
                                        # color: red, blue, green, orange,black
plt.title("Line graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
#Multiple line graphs
z=[10,12,14,16,18]

plt.plot(x,y, label="First Graph")
plt.plot(x,z, label="Second Graph")

plt.legend()
plt.show()
```

```python
plt.subplot(1,2,1) # First part refers to row, second part refers to column and third part refers to position
plt.plot([1,2,3],[4,5,6]) # first part refers to x-axis and second part refers to y-axis.

plt.subplot(1,2,2)
plt.bar(["A","B","C"],[5,3,6]) # represents bar graph

plt.show()
```

```python
#Scatter plot
plt.scatter(x,y)
plt.show()
```

```python
#horizontal bar chart
dept=["IT","HR","Manager","Admin"]
Num_emp=[30,12,10,7]
plt.barh(dept,Num_emp)
plt.figure(figsize=(20,10))
plt.show()
```

```python
#histogram
num1=[50,55,65,70,75,80,85,90]
plt.hist(num1,bins=5)
plt.show()
```

```python
marks=[45,50,55,60,65,70,75,80]
plt.hist(marks, bins=4)
plt.show()
```

```python
random_data=np.random.randint(1,51,40)
plt.hist(random_data, bins=4)

plt.show()
print(random_data)
```

```python
#pie chart
subjects=["Python","JAVA","C++","SQL","HTML"]
students=[20,15,20,20,20]

plt.pie(students,labels=subjects,autopct='%1.1f%%') # first part refers to values, second part refers to labels, autopct refers to automatic distribution of percentage
plt.figure(figsize=(5,3))
plt.show()
```
