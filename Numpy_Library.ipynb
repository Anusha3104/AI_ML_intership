```python
# creating arrays
# One dimensional array
arr1= np.array([12,32,43])

# Two dimensional array
arr2=np.array([[12,23,43],[45,11,33]])

# three dimensional array
arr3= np.array([[[1,2],[3,4]],[[5,6],[7,8]]])

print(arr1)
print(arr2)
print(arr3)
```
