```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()#Returns a list of sample datasets provided by Seaborn.
```

```python
tips = sns.load_dataset("titanic")
tips.head()
```

```python
tips.info()
```

```python
tips.describe()
```

```python
tips.shape
```

```python
sns.lineplot(x="age",y="parch",data=tips)
plt.show()
```

```python
sns.scatterplot(x="age",y="fare",data=tips)
```

```python
sns.countplot(x="sex",data=tips)#Counts how many times each category appears.
```

```python
sns.boxplot(x="sex",y="age",data=tips)#Shows the distribution of data and outliers.
```

```python
corr_mat = tips.corr(numeric_only=True)#Calculates correlation between numerical columns.
print(corr_mat)
```

```python
sns.heatmap(corr_mat)
```

```python
sns.regplot(x="fare", y="survived", data=tips)#Shows trend between two variables.
plt.show()
```
