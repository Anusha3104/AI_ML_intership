```python
#fillna(): it fills missing values Numerical data → Mean or Median
#Categorical data → Mode
df["guestroom"].fillna(df["guestroom"].mode())#mode()Returns the most frequently occurring value.


```

```python
df.drop_duplicates()#Removes duplicate rows.
```

```python
plt.boxplot(df["price"])
```

```python
#quantile() is used to find the position of values in sorted data.
q1 = df["price"].quantile(0.25)
q2 = df["price"].quantile(0.75)
iqr = q2 - q1
lower_range = q1-1.5*iqr
upper_range = q2+1.5*iqr
print(lower_range)
print(upper_range)
df = df[(df["price"] >= lower_range) &
        (df["price"] <= upper_range)]
```

```python
df=pd.get_dummies(df,drop_first=True) #Converts categorical values into numbers.
```

```python
x=df.drop('price',axis=1)#axis=1 Means column.axis=0 Means row.
y = df['price']
```

```python
model = LinearRegression()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
model.fit(x_train,y_train)
```

```python
df.select_dtypes(include='object').columns
```

```python
y_pred = model.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```
