```python
text = "I loving My , faMily!@"
```

```python
text=text.lower()
```

```python
import string
#string.maketrans:creates a translation table.This translation table tells python which characters should be removed or replaced.
#string.punctuation:this contains every punctuation symbols.
print(string.punctuation)
text = text.translate(str.maketrans('','',string.punctuation))
```

```python
#Tokenization is the process of splitting a sentence into individual words or tokens.
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt_tab')
tokens = word_tokenize(text)
print(tokens)
```

```python
#Stopwords are very common words that usually do not add much meaning.
from nltk.corpus import stopwords
nltk.download('stopwords')
stopWords= set(stopwords.words('english'))
filtered_text = [
    word for word in tokens if word.lower() not in stopWords
]
```

```python
#Stemming reduces words to their root form by removing prefixes or suffixes using simple rules.
from nltk.stem import PorterStemmer

ps = PorterStemmer()
for word in filtered_text:
    print(ps.stem(word))
```

```python
#Lemmatization:Lemmatization converts a word into its actual dictionary (base) form using vocabulary and grammar.
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemma = WordNetLemmatizer()

lemma.lemmatize("loving", pos='v')
```
